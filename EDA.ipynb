## Deep Exploration on Specific Categorical Feature vs Price

Lets say we want to look at the price distribution based on certain categorical feature, to see how the price distribution of different types of specifications.

**Example:**
* View price distribution by drive-wheels (FWD, RWD, 4WD)

We can use a boxplot to see this

```python
#Create visualization
sns.boxplot(x='drive-wheels', y='price', data=df, palette='Set2')
```

**Interesting Findings:**
* Rear-drive (RWD) cars have the widest price variation, indicating that this type of drive is likely to be found on a wide range of cars, from cheap to luxury.
* Front- and 4-wheel drive cars (FWD & 4WD) are concentrated in low-priced cars (<15,000), with a narrower price range for 4WD in particular.
* Outliers are seen across all drive types, indicating each drive is used by a few cars above the majority price range.

**Deep Dive More**

After seeing the price distribution per drive-wheels, 

we can deep dive again by adding categorical features such as body-style and see the average price for each car based on drive-wheels and body-style.

```python
#Specify the categorical features to be used
df_test = df[['drive-wheels', 'body-style', 'price']]

#Grouping the data
df_grp = df_test.groupby(['drive-wheels', 'body-style'], as_index=False).mean() #price mean per drive-wheels and body-stle

#Pivot format
df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
```

**Visualize the pivots**

```python
#Create visualization
plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.xticks(np.arange(df_pivot.shape[1]) + 0.5, df_pivot.columns.levels[1], rotation=0)
plt.yticks(np.arange(df_pivot.shape[0]) + 0.5, df_pivot.index)

#Define the x and y label
plt.xlabel("Body Style")
plt.ylabel("Drive Wheels")

#Show the plot
plt.show()
```

## Explore the Correlation between Numerical Feature vs Price

```python
#Calculating IQR to detect outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Filtered outlier data
df_filtered = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

#Define the numerical feature to be use
num_var = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'highway_L/100km', 'city_L/100km']

#Create visualization
fig, axes = plt.subplots(2, 7, figsize=(35, 15))
for num, ax in zip(num_var, axes.flatten()):
    sns.regplot(data=df_filtered, x=num, y='price', ax=ax)

plt.tight_layout()
plt.show()
```

```python
#Specify the numerical features to be used
data_test = df[['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'highway_L/100km', 'city_L/100km', 'price']]
data_test_filtered = data_test[(data_test['price'] >= lower_bound) & (data_test['price'] <= upper_bound)]
data_corr = data_test_filtered.corr()

# Create visualization
plt.figure(figsize=(12, 8))

# Heatmap
sns.heatmap(data_corr, annot=True, fmt='.2f')
```

**Interesting Findings:**

* In general, there is a positive correlation between most car features (such as width, length, engine size, horsepower, etc.) and car prices. That is, the higher the value of these features, the more expensive the car tends to be.
* Lenghth, width, curb-weight, engine-size, horsepower, fuel consumption (L/100km) are the specifications that have a strong correlation with price, indicating that the higher these specifications are, the more expensive the car is.
* Interestingly, fuel consumption (L/100km) has a positive correlation. This indicates that the existing data has expensive cars that fall into the sports/performance category not just luxury, with high performance, fuel consumption is also high.
* Height, stroke, compression ratio, and peak rpm do not have a strong enough correlation with price, especially peak rpm.
